```python
import pandas as pd
import numpy as np

np.random.seed(500)
```

```python
df = pd.read_csv('./data/mimic-iii-clinical-database-demo-1.4/PRESCRIPTIONS.csv', dtype=str)
df
```

```python
df_ndc_freq = df[['row_id', 'ndc']].groupby('ndc').count().sort_values('row_id',ascending=False)
df_ndc_freq
```

```python
df_ndc_freq = df_ndc_freq.reset_index()
```

```python
df_ndc_most_freq_codes = df_ndc_freq[(df_ndc_freq['row_id'] >= 50) & (df_ndc_freq['ndc'] != "0")]
df_ndc_most_freq_codes
```

```python
most_freq_ndc_codes = df_ndc_most_freq_codes['ndc'].to_list()
```

```python
df_ndc_most_freq_meta = df[df['ndc'].isin(most_freq_ndc_codes)]
df_ndc_most_freq_meta
```

```python
df_ndc_most_freq = df_ndc_most_freq_meta[['ndc','drug','drug_name_poe','drug_name_generic']].drop_duplicates()
df_ndc_most_freq
```

```python
ndc_list = df_ndc_most_freq.to_dict(orient='list')
```

```python
drug_code_names_map = {}

ndc_codes = ndc_list['ndc']
drugs = ndc_list['drug']
drug_name_poes = ndc_list['drug_name_poe']
drug_name_generics = ndc_list['drug_name_generic']

number_of_records = len(ndc_codes)

for i in range(number_of_records):
    ndc_code = ndc_codes[i]
    drug = drugs[i]
    drug_name_poe = str(drug_name_poes[i])
    drug_name_generic = str(drug_name_generics[i])

    if ndc_code not in drug_code_names_map:
        drug_code_names_map[ndc_code] = set()

    drug_code_names_map[ndc_code].add(drug)

    if drug_name_poe != 'nan':
        drug_code_names_map[ndc_code].add(drug_name_poe)
    
    if drug_name_generic != 'nan':
        drug_code_names_map[ndc_code].add(drug_name_generic)

```
